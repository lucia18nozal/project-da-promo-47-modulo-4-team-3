```python
from kaggle.api.kaggle_api_extended import KaggleApi

import os
print(os.path.exists(os.path.expanduser("~/kaggle.json")))  #if false, it can't find the authentication key.
os.environ["KAGGLE_CONFIG_DIR"] = "C:\\Users\\Kathe\\.kaggle"
import pandas as pd

api = KaggleApi()
api.authenticate()
```

```python
datasets = api.datasets_list(search="World Happiness Report")
print(datasets)
```

```python
len(datasets) #to see how many lists came up in my search
```

```python
df_datasets = pd.DataFrame(datasets)
pd.set_option('display.max_columns', None)
```

```python
df_datasets.rename(columns={'ref': 'slug'}, inplace=True) #this is the column I need to reference for downloading datasets
print(df_datasets.columns.get_loc('slug')) #column indexfor slug value

df_datasets #to better visualize the results of the search in Kaggle
```

```python
# Specify the dataset slug (use the slug from the URL)
dataset_slug = 'dataset-owner/dataset-name'
download_path = r"C:\Users\Kathe\Documents\source_code\ejercicios-en-adalab\Proyecto Personal\csv_proyecto_personal"


def get_datasets(dataset_slug):
    try:
        api.dataset_download_files(dataset_slug, path=download_path, unzip=True)
        print(f"Dataset downloaded and extracted to {download_path}")

    except Exception as e:
        print(f"Error downloading dataset: {e}")
```
