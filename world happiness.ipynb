```python
lista_nulos = df2.columns[df2.isnull().sum() > 0] #columnas con valores nulos

lista_nulos
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3, figsize = (20,10)) 
axes = axes.flat
for indice, col in enumerate(lista_nulos):
    sns.boxplot(x = col, data = df2, ax = axes[indice])
    
plt.tight_layout()
fig.delaxes(axes[-1]);
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 10))
axes = axes.flat  # Flatten the 3x3 grid into a 1D array

for indice, col in enumerate(lista_nulos):
    axes[indice].hist(df2[col].dropna(), bins=30, color='skyblue', edgecolor='black')
    axes[indice].set_title(f'Histograma de {col}')
    axes[indice].set_xlabel(col)
    axes[indice].set_ylabel('Frecuencia')

plt.tight_layout()  # Adjust layout to prevent overlapping
fig.delaxes(axes[-1])  # Remove the last empty subplot if necessary
plt.show()


#todos tienen outliers y no vemos una distribución simetrica, por lo que imputaremos con la mediana
```

```python
df2[lista_nulos] = df2[lista_nulos].fillna(df2[lista_nulos].median())

df2.isnull().sum() #comprobar que ha imputado todos los nulos
```

```python
df = pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
df
```

```python
df['Country name'].value_counts()[df['Country name'].value_counts() == 1]
 #clos paises que solo están en uno de los dos datasets
```

```python
df['Country name'] = df['Country name'].str.replace('Czechia', 'Czech Republic')
df['Country name'] = df['Country name'].str.replace('Palestinian Territories', 'State of Palestine')
df['Country name'] = df['Country name'].str.replace('Turkiye', 'Turkey')
df['Country name'] = df['Country name'].str.replace('Eswatini', 'Swaziland')
```

```python
df_2021_2024 = pd.read_csv('files/world_happiness_report_2021_2024.csv')
df_2021_2024
```

```python
#vamos a explorar los datos de GDP para 2021 y 2024 por pais

df_gdp = df_2021_2024[['Country name', 'year', 'Log GDP per capita']]

df_gdp
```

```python
# Filter the data for 2021 and 2024
df_gdp = df_2021_2024[df_2021_2024['year'].isin([2021, 2024])]

# Pivot the table to have separate columns for 2021 and 2024
df_gdp_pivot = df_gdp.pivot(index='Country name', columns='year', values='Log GDP per capita')

# Rename the columns for clarity
df_gdp_pivot.columns = ['GDP 2021', 'GDP 2024']

# Reset the index to make it a clean DataFrame
df_gdp_pivot = df_gdp_pivot.reset_index()

# Show the result
print(df_gdp_pivot)
```
